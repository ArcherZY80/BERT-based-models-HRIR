```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd
import logging
from matplotlib import pyplot as plt

logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```python
from scipy.special import softmax
from google.colab import files
```

#Data preparation

```python
train_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/project/4192/Data/DirectCompare/train_df.csv")

train_df = train_df[['selftext','Expert-label']]
train_df
```

```python
test_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/project/4192/draft/Iteration/simualte_hrir/Iteration1/bert-base/b_b_i1.csv")

test_df = test_df[['selftext']]
test_df
```

```python
test_data = test_df['selftext'].values.tolist()
```

#Model

```python
model_args = ClassificationArgs(train_batch_size=4, gradient_accumulation_steps=16, learning_rate=3e-5, num_train_epochs= 3, max_seq_length= 512, overwrite_output_dir= True)
```

##roberta-base

```python
model_bt = ClassificationModel('roberta', 'roberta-base',  args = model_args, use_cuda = True)
model_bt.train_model(train_df)
```

##deberta-base

```python
model_db = ClassificationModel('deberta', 'microsoft/deberta-base',  args = model_args, use_cuda=True)
model_db.train_model(train_df)
```

#Validate

```python
result_bt,proba_bt = model_bt.predict(test_data)
probabilities = softmax(proba_bt, axis=1)
proba_bt_df=pd.DataFrame(probabilities)
proba_bt_df.columns = ["remove","Expert-label"]
proba_bt_df = proba_bt_df["Expert-label"]
proba_bt_df
```

```python
result_db,proba_db = model_db.predict(test_data)
probabilities = softmax(proba_db, axis=1)
proba_db_df=pd.DataFrame(probabilities)
proba_db_df.columns = ["remove","Expert-label"]
proba_db_df = proba_db_df["Expert-label"]
proba_db_df
```

```python
val_df=proba_bt_df+proba_db_df
val_df
```

```python
val_df=val_df.to_frame()
val_df
```

```python


val_df.loc[val_df['Expert-label'] < 0.9] = 0
val_df.loc[val_df['Expert-label'] >= 0.9] = 1
val_df
```

```python
save_df = pd.concat([test_df,val_df],axis=1)
save_df.head(200)
```

```python
#Saving to drive

save_df.to_csv('b_b_val_i1.csv')
files.download('b_b_val_i1.csv')

```
